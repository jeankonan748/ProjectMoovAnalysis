```python
import pandas as pd

df = pd.read_excel("DatasetCallCenter - 2026-05-05T082442.350.xlsx")

# Conversion des dates
df["Créé le"] = pd.to_datetime(df["Créé le"])
df["Date de Clôture"] = pd.to_datetime(df["Date de Clôture"])

# Création du DMT (durée de traitement)
df["DMT"] = df["Date de Clôture"] - df["Créé le"]
```

```python
df
```

# Feature engineering

```python
df["Jour"] = df["Créé le"].dt.day_name()
df["Heure"] = df["Créé le"].dt.hour
df["Mois"] = df["Créé le"].dt.month
df
```

# ANALYSES CLÉS

```python
print(df["DMT"].describe())
```

# DMT par priorité

```python
dmt_priorite = df.groupby("Priorité")["DMT"].mean().sort_values()
print(dmt_priorite)
```

# DMT par file d’attente

```python
dmt_file = df.groupby("File d'attente")["DMT"].mean().sort_values(ascending=False)
print(dmt_file)
```

# Volume de tickets

```python
volume = df["Créé le"].dt.date.value_counts().sort_index()
print(volume)
```

# Performance agents

```python
perf_agents = df["Créé par"].value_counts()
print(perf_agents)
```

# MACHINE LEARNING

## Prédire le DMT

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# Sélection de features simples
X = df[["Priorité"]]
X = pd.get_dummies(X)

y = df["DMT"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)

print("Score:", model.score(X_test, y_test))
```
